```python
C_index = d['pC'].size
E_index = d['pE'].size
if d['pC'].size == 0:
    bpE_last = 400 
    first_C = 300
    lin_C = range(first_C,bpC_last,1)
elif d['pC'].size != 0:
    bpC_last = int(d['pC'][C_index-1])
    first_C = int(d['pC'][1])
    lin_C = range(first_C,bpC_last,1)
if d['pE'].size == 0:
    bpE_last = 400 
    first_E = 300
    lin_E = range(first_E,bpE_last,1)
elif d['pE'].size != 0:
    bpE_last = int(d['pE'][E_index-1])
    first_E = int(d['pE'][1])
    lin_E = range(first_E,bpE_last,1)
#print(bpC_last)
#print(bpE_last)
nb_comb = int(len(range(0,bpE_last,1))*len(range(0,bpC_last,1)));
#print(nb_comb)
```

```python
#error_stored = np.zeros(nb_comb//1)
#bp_stored = np.zeros((nb_comb//1,2))
bp_couple = np.zeros((nb_comb//1,2))
mean_error = []
a=0
for i in tqdm(lin_C) :
    for j in lin_E:
#for i in range(0,bpC_last,1):
#    for j in range(0,bpE_last,1):
        bp_couple[a,0] = i
        bp_couple[a,1] = j
        P1, P2 = create_P1_and_P2(data,i,j)
        P1_coeff = [P1.A,P1.B,P1.C]
        P2_coeff = [P2.A,P2.B,P2.C]
        sP1 = np.dot(np.transpose(P1.sig123),P1_coeff)
        sP2 = np.dot(np.transpose(P2.sig123),P2_coeff)
        errP1 = abs(1-sP1)
        errP2 = abs(1-sP2)
        sumP1 = np.sum(errP1)
        sumP2 = np.sum(errP2)
        mean_error.append((sumP1 + sumP2)/(len(errP1)+len(errP2)))
        a=a+1
        
mean_error = np.array(mean_error)
print(mean_error.tolist())
min_error_index = np.nanargmin(mean_error)
print(min_error_index)
print(min(mean_error))
#print(mean_error[])
```
